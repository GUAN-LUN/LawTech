```python
#靜態爬取資料
from time import sleep        # 爬蟲禮貌，避免對方伺服器無法負荷請求而宕機
import requests               # 爬蟲請求
import json                   # JSON 資料型態
import pandas as pd  # 資料處理套件
from bs4 import BeautifulSoup

first_clean_list=[]
number=[]
date=[]
reason=[]
content=[]

for i in range(0,5):
    #查找網站
    try:
        a = str(i).format()
        res = requests.get(f'https://law.judicial.gov.tw/FJUD/data.aspx?ro={a}&q=ae20195856474492de9ee2ed7ce23b41&sort=DS&ot=in')
        soup = BeautifulSoup(res.text , 'html.parser')

        target_table = soup.find_all(class_='col-td')
        #刪除裁定資料
        if "裁定" not in target_table[0].text:
            print(target_table[0])
            number.append(target_table[0].text)
            date.append(target_table[1].text)
            reason.append(target_table[2].text)
            content.append(target_table[3].text)
    except: pass

    


#轉成dict
law_array=[]
for i in range(len(number)):
    law_array.append({
                            "裁判字號": number[i],
                            "裁判日期": date[i],
                            "裁判案由":reason[i],
                            "內文":content[i]})
law_df = pd.DataFrame.from_dict(law_array)

#輸出csv
law_df.to_csv('law_final_land.csv', encoding='utf-8')
```
